### Hsp67Bc gene expression in Drosophila melanogaster larva

##### Source of the data: http://flybase.org/reports/FBgn0001229

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Source
fly = pd.read_csv("FBgn0001229G.csv", sep=";", index_col=None)
fly = fly.sort_values(by="FPKM")
```

```python
f= 13
plt.figure(figsize=(8, 3))
plt.title("Gene expression of Hsp67Bc in larva - results from RNAseq",fontsize=f )

plt.barh(list(fly["Tissue"]), list(fly["FPKM"]), xerr=list(fly["SD"]), color="wheat")
plt.ylabel("Tissue",fontsize=f)
plt.xlabel("Fragments per kilobase of exon model per million mapped reads",fontsize=f)
plt.savefig("expression-in-larva")
```
